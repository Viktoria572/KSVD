# Яременко Вікторія Лабораторна 2

Побудувати двонаправлену RNN для розрахунку функції із шумом.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Генеруємо дані
N = 10000
X = np.random.rand(N) * 10  # генеруємо N точок від 0 до 10
Y = np.cos(3*X) + 0.1 * np.random.randn(N)  # вхідні дані - косинус з шумом

# Побудова моделі нейронної мережі
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, input_shape=(1,), activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Компіляція моделі
model.compile(optimizer='adam', loss='mse')

# Навчання моделі та збереження історії навчання
history = model.fit(X, Y, epochs=10, batch_size=32, verbose=1)

# Передбачення за допомогою моделі
predictions = model.predict(X)

# Побудова графіків
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, Y, label='Дані')
plt.scatter(X, predictions, color='r', label='Прогнози')
plt.title('Співставлення Даних та Прогнозів')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title('Функція Втрат')
plt.xlabel('Епохи')
plt.ylabel('Втрати')

plt.show()
```
