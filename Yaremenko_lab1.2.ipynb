```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантаження даних з CSV файлу
df = pd.read_csv('/content/sample_data/shopping_trends (1).csv', encoding='latin-1')

# Перевірка розміру таблиці
print("Розмір таблиці:", df.shape)

# Перевірка наявності пропусків
print("Наявність пропусків:\n", df.isnull().sum())

# Перевірка наявності дублікатів
print("Кількість дублікатів:", df.duplicated().sum())

# Отримання інформації про типи стовпців
print("Типи стовпців:\n", df.dtypes)

# Перейменування назв стовпців
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Вивід першіх рядків для перевірки
print("Перші рядки з оновленими назвами стовпців:\n", df.head())
```

Побудуйте графіки з використанням бібліотеки seaborn.

1. Гістограма віку (age):

```python
# Побудова гістограми віку
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=False, color='skyblue')
plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

2. Порівняння розподілу покупок за статтею (item_purchased):

Створіть стовпчасту діаграму або графік розподілу, щоб порівняти, які товари покупають клієнти найчастіше.

```python
# Побудова стовпчастої діаграми розподілу покупок за статтею
plt.figure(figsize=(12, 6))
sns.countplot(x='item_purchased', data=df, palette='Set2', order=df['item_purchased'].value_counts().index)
plt.title('Порівняння розподілу покупок за статтею')
plt.xlabel('Товар')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

3. Залежність між віком та сумою покупок (purchase_amount):

Використовуйте scatter plot, щоб візуалізувати залежність між віком та сумою покупок.

```python
# Побудова scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df, color='coral')
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок (USD)')
plt.show()
```

7. Сезонність покупок (season):

8. Графік розподілу розмірів одягу (size):
